## fifa_df19 = players_19.csv에서 sofifa_id','short_name','overall' 컬럼만 추출
## fifa_df20 = players_20.csv 에서 sofifa_id와 overall만을 추출 
## sofifa_id를 key값으로  위의 두 테이블을 innerjoin 하세요 

```python
# fifa_df19 = players_19.csv에서 sofifa_id','short_name','overall' 컬럼만 추출
data1 = fifa_df19[['sofifa_id','short_name','overall']]
data1
```

```python
# fifa_df20 = players_20.csv 에서 sofifa_id와 overall만을 추출
data2 = fifa_df20[['sofifa_id','overall']]
data2
```

```python
# sofifa_id를 key값으로 위의 두 테이블을 innerjoin 하세요
inner = pd.merge(data1,data2,how = 'inner',on = 'sofifa_id')
inner
```

## sofifa_id를 key값으로  위의 두 테이블을 outerjoin 하세요 

```python
outer = pd.merge(data1, data2, how = 'outer', on = 'sofifa_id')
outer
```

## outerjoin을 한 테이블 중 inner join을 한 테이블과 중복되는 행을 삭제하세요 

```python
outer.loc[~outer['sofifa_id'].isin(inner['sofifa_id'])]
```
