```python
import pandas as pd
df = pd.read_csv('./data/7.세종시_아파트(전월세)_실거래가.csv')
df
```

```python
#1) 세종시 데이터에서 컬럼 확인 후, unnamed 컬럼 제거
df.drop(columns = 'Unnamed: 0',inplace = True)
df
```

```python
#2) 전월세구분 컬럼에서 unique값 확인  
df['전월세구분'].unique()
```

```python
#3) index= 100~200, column = 전월세구분, 전용면적(㎡),보증금(만원)으로 이루어진 데이터프레임 추출
df.loc[100:200,['전월세구분','전용면적(㎡)','보증금(만원)']]
```

```python
###4) 0~200 index를 가진 행 삭제 
df.drop(df.index[0:201])
```

```python
#5) 건축년도 컬럼 중복 제거 결과 출력  
df.건축년도.drop_duplicates().sort_values()
```
