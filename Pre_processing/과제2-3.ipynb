    데이터 : "7.세종시_아파트(전월세)_실거래가.csv"
    1) 단지명 중  '해밀마을 2단지'인 row 삭제 
    2) new_시군구' 컬럼을 만들고 '세종특별자치시 고운동' => '세종특별자치시 해밀동' 으로 변경
    3) 'Class'컬럼을 만든 뒤 전세면 True 아니면 False로 입력 (loc사용) 
    4) 전세이면서 보증금이 1억 이상인 데이터만을 추출 
    5) -10~0층 : 지하 , 1~5층 : 저층 6~15층 : 중층, 16층 이상 : 고층으로 할당하는 새로운 컬럼 생성 

```python
import pandas as pd
df = pd.read_csv('./data/7.세종시_아파트(전월세)_실거래가.csv')
df.drop(columns = 'Unnamed: 0',inplace = True )
df
```

```python
# 1) 단지명 중  '해밀마을 2단지'인 row 삭제 
ans1 = df.loc[df['단지명'] != '해밀마을 2단지']
ans1
```

```python
# 2) new_시군구' 컬럼을 만들고 '세종특별자치시 고운동' => '세종특별자치시 해밀동' 으로 변경
df.loc[df['시군구'] == '세종특별자치시 고운동','new_시군구'] = '세종특별자치시 해밀동'
df
```

```python
# 3) 'Class'컬럼을 만든 뒤 전세면 True 아니면 False로 입력 (loc사용) 
df.loc[df['전월세구분'] == '전세' ,'Class'] = True
df.loc[df['전월세구분'] != '전세', 'Class'] = False
df
```

```python
# 4) 전세이면서 보증금이 1억 이상인 데이터만을 추출 
df.loc[(df['전월세구분'] == '전세') & (df['보증금(만원)'] >= 10000)]
```

```python
# 5) -10~0층 : 지하 , 1~5층 : 저층 6~15층 : 중층, 16층 이상 : 고층으로 할당하는 새로운 컬럼 생성 
import numpy as np
df['층구분'] = np.where(df['층']<=0, '지하', np.where(df['층']<=5,'저층', np.where(df['층']<=15,'중층', np.where(df['층']>=16,'고층',None))))
df['층구분'].value_counts()
```
