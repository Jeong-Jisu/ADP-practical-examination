# 과제 3-3 문자열 데이터 변환하기 (str) 

## 1 '6.세종시_오피스텔(매매)_실거래가.csv' 데이터 불러오기 


```python
import pandas as pd
df = pd.read_csv('../data/6.세종시_오피스텔(매매)_실거래가.csv')
df
```

## 2. 시군구 띄어쓰기로 3개의 컬럼으로 변경 


```python
df[['시','군','구']]=df['시군구'].str.split(' ',expand = True)
df
```

## 3. 도담동 => 고운동으로 변경 


```python
df.loc[df['군'] == '도담동','군'] = '고운동'
df['군'].unique()
```

## 4. 세종시 오피스텔 거래금액(만원)을 int형으로 변경하세요 


```python
df['거래금액(만원)'] = df['거래금액(만원)'].str.split(',').str.get(0) + df['거래금액(만원)'].str.split(',').str.get(1)
df['거래금액(만원)'] = df['거래금액(만원)'].astype(int)
df['거래금액(만원)']
```

## 5. 시군구별 전용면적(㎡)별 오피스텔 거래금액 평균 값 구하기 


```python
df.groupby(['시군구'])[['전용면적(㎡)']].mean()
```

## 6. 계약년월 별, 평균 오피스텔 가격 구하기  

```python
df.groupby(['계약년월'])[['거래금액(만원)']].mean()
```
