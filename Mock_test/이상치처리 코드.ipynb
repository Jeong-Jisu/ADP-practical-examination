```python
# 일부 변수들에서 이상치가 확인되며 전처리가 필요할 것으로 보인다. 
# LotArea, LotFrontage, 1stFlrSF,2ndFlrSF,price의 이상치를 처리하고자 한다.

fig, axes = plt.subplots(2, 7, figsize = (12,12))
for ax, col in zip(axes.ravel(),df.columns):
    ax.boxplot(df[[col]])
    ax.set_xlabel(col)
plt.show()
```

```python
'''
ID 는 분석 상 별 의미가 없어 삭제하였고, 결측치가 있는 LotFrontage의 결측치를 KNNImputer로 대치하였다.
YearBuilt와 YearRemodAdd는 순서가 있는 범주형 변수인데 각 연도를 그대로 쓰기 보다 binning 전처리를 함으로써
더 유의미한 변수가 되고 계산비용도 줄이고자 하였다. 마지막으로 이상치를 조정함으로써 왜곡된 학습결과를 피하고자 하였다.

'''
df2 = df.drop('Id', axis = 1)
from sklearn.impute import KNNImputer
imputer = KNNImputer()
df2 = pd.DataFrame(imputer.fit_transform(df2), columns = df2.columns)

df2.loc[df2['YearBuilt'] <= 1960,'YearBuilt'] = 1
df2.loc[(df2['YearBuilt'] > 1960) & (df2['YearBuilt'] <= 1970),'YearBuilt'] = 2
df2.loc[(df2['YearBuilt'] > 1970) & (df2['YearBuilt'] <= 1980),'YearBuilt'] = 3
df2.loc[(df2['YearBuilt'] > 1980) & (df2['YearBuilt'] <= 1990),'YearBuilt'] = 4
df2.loc[(df2['YearBuilt'] > 1990) & (df2['YearBuilt'] <= 2000),'YearBuilt'] = 5
df2.loc[(df2['YearBuilt'] > 2000) & (df2['YearBuilt'] <= 2010),'YearBuilt'] = 6

df2
```

```python
for column in ['price']:
    IQR = df2[column].quantile(0.75) - df2[column].quantile(0.25)
    top = df2[column].quantile(0.75) + IQR * 1.5
    down = df[column].quantile(0.25) - IQR * 1.5
    df2[column] = df2.clip(down,top)[column]
```

```python
fig, axes = plt.subplots(2, 7, figsize = (12,12))
for ax, col in zip(axes.ravel(),df2.columns):
    ax.boxplot(df2[[col]])
    ax.set_xlabel(col)
plt.show()
```
