```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/adp/24/problem2.csv',encoding = 'cp949')
df
```

# 1. 광고비 변수를 가변수 처리후 다중회귀를 수행하여 회귀계수가 유의한지 검정하시오.


```python
# 범주형 변수인 광고비를 get_dummies로 더미코딩 하는 전처리를 진행하였다. 
# 범주의 수준보다 1개의 개수가 적은 가변수를 생성함으로써 다중공선성을 피할 수 있다.
df2 = pd.get_dummies(df, columns = ['광고비'], drop_first = True)
df2
```

```python
'''
회귀계수의 유의성을 위해 회귀계수가 0인지 0이 아닌지에 대해 t-검정을 실시한다. 
t-검정에 대한 귀무가설은 회귀계수는 0과 차이가 없다이고, 대립가설은 회귀계수가 0과 차이가 있다. 즉, 
회귀계수가 통계적으로 유의하다이다.

statsmodels를 통해 아래와 같이 각 회귀계수를 구하고, t-검정을 실시한 결과를 구하였다.
광고횟수의 회귀계수는 1.4350, t 통계량은 19.518이며 이 때 p-val는 0.000을 나타낸다.
유의수준 보다 p-value가 작기 때문에 귀무가설을 기각하고 대립가설을 채택한다. 즉, 광고횟수의 회귀계수는 통계적으로 유의하다.
광고비의 회귀계수 0.5227이고, t통계량은 0.975이며 p-value는 0.358이다. p-value가 유의수준보다 크기 때문에 귀무가설을 기각할 수 없으며 해당 회귀계수는 통계적으로 유의하다고 볼 수 없다
'''

import statsmodels.api as sm
X = df2[['광고횟수', '광고비_높음']]
y = df2['매출액']
X_ = sm.add_constant(X)
model = sm.OLS(y,X_)
fitted = model.fit()
fitted.summary()
```

# 회귀식이 유의한지 판단하시오.

* 회귀식의 유의성은 다중회귀 모델의 F 통계량을 통해 판단한다. 앞에서 얻은 F통계량은 221.2이고, 그 p-value는 유의수준인 0.05보다 작다. 이를 근거로 해당 회귀식은 통계적으로 유의하다고 할 수 있다. 

# 모델의 설명력(예측 정확도)은 어떤가?

* 모델의 설명력은 수정된 R^2으로 알 수 있다. 수정된 R^2의 범위는 0~1까지이며, 모형의 예측 정확도가 높아지면 결정계수가 1에 가까워짐을 알 수 있다. 
* 현재 모델의 수정된 R^2을 살펴보면 0.978로 상당히 높은 설명력을 가지고 있다고 할 수 있다
